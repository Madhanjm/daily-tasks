```python
import matplotlib.pyplot as plt
matches_per_season = df['season'].value_counts().sort_index()


plt.plot(matches_per_season.index, matches_per_season.values, marker='o')
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.title("IPL Matches per Season")

plt.grid()
plt.show()
```

```python
team_wins=df['winner'].value_counts()

plt.bar(team_wins.index,team_wins.values)
plt.xlabel("Teams")
plt.ylabel("Wins")
plt.xticks(rotation=90)

plt.title("Total win by each team")
plt.show()
```

```python
toss_decision = df['toss_decision'].value_counts()

plt.figure()
plt.pie(toss_decision, labels=toss_decision.index, autopct="%1.1f%%",explode=(0.1,0.1))
plt.title("Toss Decision Distribution")
plt.show()
```
