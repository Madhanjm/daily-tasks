```python
import pandas as pd
data = {
    'patient_id':[1, 2, 3, 1, 2, 3, 4, 5],
    'timestamp':['2024-01-01 07:00', '2024-01-01 08:00', '2024-01-01 07:00', 
                    '2024-01-02 08:00', '2024-01-02 09:00', '2024-01-02 10:00', 
                    '2024-01-01 11:00', '2024-01-01 01:00'],
    'heart_rate':[70, 80, 95, 120, 85, 76, 65, 130],
    'ward': ['A', 'B', 'A', 'A', 'B', 'A', 'B', 'A']
}

df=pd.DataFrame(data)
print(df)
```

```python
#Compute average heart rate per ward
avghr_perward=df.groupby('ward')['heart_rate'].mean()
print(avghr_perward)
```

```python
#Identify patients with sudden spikes in heart rate
df=df.sort_values(by=['patient_id','timestamp'])

df["hrt_rate_diff"]=df.groupby('patient_id')['heart_rate'].diff()

sudden_spk=df[df['hrt_rate_diff']>25]
print("sudden spike : ",sudden_spk[['patient_id','timestamp','hrt_rate_diff']])
```

```python
#Extract data for critical cases
critical_case=df[df['heart_rate']>100]
print("critical case",critical_case[['patient_id','timestamp','heart_rate','ward']])
```
